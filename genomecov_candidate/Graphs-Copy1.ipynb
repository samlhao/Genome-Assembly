# Imports

```python
%matplotlib inline    # automatically plots in jupyter notebooks
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Read Data

```python
# As an examplt, we have these files with some gene IDs to look for
files = ["881_genomecov.tsv", "882_genomecov.tsv",
         "887_genomecov.tsv", "888_genomecov.tsv",
         "889_genomecov.tsv", "890_genomecov.tsv"]
IDs = ['arn']
letters = ['A', 'B', 'C', 'D']

# Make a dictionary of pandas DataFrames with filenames as keys
df_dict = {}
for f in files:
    df = pd.read_csv(f, sep='\t', names=['text', 'index', 'value']) #names are for the referencing
    df_dict[f] = df
```

# A Few Plotting Examples

```python
# Here we just plot them one by one, in the desired order
# Note: this is quite slow --- always try to avoid using for loops in python 
#     in order to take advantage of the speed of libraries like numpy that are 
#     precompiled. I didn't know off the top of my head how to search for the 
#     keyword "current_gene" in a pandas dataframe, but I bet you can and then you 
#     don't need to do the loop at the end that takes forever
for ID in IDs:
    for let in letters:
        current_gene = ID + let
        for file in files:
            f, ax = plt.subplots(1, figsize=(10,7))
            ax.set_title(current_gene + ": " + file)
            ax.set_xlabel('postion')
            ax.set_ylabel('coverage')
            df = df_dict[file]
            position = [] #BAD
            coverage = [] #BAD
            for i in range(len(df)): #BAD
                if current_gene in df['text'][i]:
                    position.append(df['index'][i]) #BAD
                    coverage.append(df['value'][i]) #BAD
            ax.plot(position, coverage)
```

```python
# Here we plot all the same genes in one plot with colors showing 
# the different files of origin
for ID in IDs:
    for let in letters:
        current_gene = ID + let
        f, ax = plt.subplots(1, figsize=(10,7))
        ax.set_title(current_gene)
        ax.set_xlabel('postion')
        ax.set_ylabel('coverage')
        for file in files:
            df = df_dict[file]
            position = []
            coverage = []
            for i in range(len(df)):
                if current_gene in df['text'][i]:
                    position.append(df['index'][i])
                    coverage.append(df['value'][i])
            ax.plot(position, coverage, label=file) #label is for the legend. we use the file name
        ax.legend()
```
